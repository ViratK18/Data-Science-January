```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df
```

```python
sns.relplot(x='total_bill',y='tip',data=df)
```

```python
#Now we want to differentiate with sex

sns.relplot(x='total_bill',y='tip',data=df,hue='sex')
```

```python
#Now we want to differentiate it with time too

sns.relplot(x='total_bill',y='tip',data=df,hue='sex',style='time')
```

```python
#Now we want to differentiate it with size too

sns.relplot(x='total_bill',y='tip',data=df,hue='sex',style='time',size='size')
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df,hue='sex',style='time',size='size')
```

```python
#We convert relplot to lineplot

sns.relplot(x='total_bill',y='tip',data=df,hue='sex',style='time',size='size',kind='line')
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='smoker',col='time')
```

##Categorical Plot

```python
sns.catplot(x='day',y='total_bill',data=df)
```

```python
sns.catplot(x='day',y='total_bill',data=df,jitter=False)
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind='swarm')
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind='swarm',hue='size')
```

```python
sns.catplot(x='smoker',y='tip',data=df,kind='swarm')
```

```python
sns.catplot(x='smoker',y='tip',data=df,kind='swarm',order=['No','Yes'])
```

```python
#Make boxplot with the help of catplot

sns.catplot(x='day',y='tip',data=df,kind='box')
```

```python
sns.catplot(x='day',y='tip',data=df,kind='boxen')
```

```python
sns.catplot(x='total_bill',y='day',data=df,hue='sex',kind='violin')
```

```python
#split is by default false
sns.catplot(x='total_bill',y='day',data=df,hue='sex',kind='violin',split='true')
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind='violin',inner=None)
```

```python
sns.swarmplot(x='day',y='total_bill',data=df,color='k',size=3)
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind='violin',inner=None)
sns.swarmplot(x='day',y='total_bill',data=df,color='k',size=3)
```
